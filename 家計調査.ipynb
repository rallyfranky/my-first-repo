```python
from google.colab import drive
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/drive/MyDrive/年齢別消費支出.csv",encoding="SHIFT-JIS",index_col="用途分類")
```

```python
df.head()
```

```python
for i ,col in enumerate(df.columns):
  df.iloc[:,i] = df.iloc[:,i].astype(str)
df.info()
```

```python
sc = StandardScaler()
df_cluster = sc.fit_transform(df)
df_cluster
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
 
linkage_result = linkage(df_cluster, method='ward', metric='euclidean')
plt.figure(num=None, figsize=(9, 9),dpi=120,facecolor='w', edgecolor='k')
dendrogram(linkage_result, labels=df.index)
plt.show()
```
