```python
#ロンドンデータセットの取得

import pandas as pd
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
target
```

```python
data.shape
```

```python
#カラム名を設定、データフレームの形式にして取得
column_names = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
df = pd.DataFrame(data,columns=column_names)
```

```python
df['y']=target
```

```python
df
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data = df.loc[:,['RM','LSTAT','y']]
data.head()
```

```python
plt.hist(data['y'])
```

```python
plt.scatter(data['RM'],data['y'])
```

```python
plt.scatter(data['LSTAT'],data['y'])
```

```python
#RMのみを説明変数としてNormで最尤推定
result = smf.glm(formula='y ~ RM',data=data,family=sm.families.Gaussian()).fit(disp=0)
result.summary()
```

```python
#予測と実績値の比較
plt.scatter(data['RM'],data['y'],label='accutual')
plt.scatter(data['RM'],result.predict(),marker='x',label='predict')
plt.legend()
plt.show()
```

```python
#他モデルとの比較のためmseを算出
from sklearn.metrics import mean_squared_error
mean_squared_error(data['y'],result.predict())
```

```python
result = smf.glm(formula='y ~ RM+LSTAT',data=data,family=sm.families.Gaussian()).fit(disp=0)
result.summary()
```

```python
# LSTAT={5, 15, 25, 35}の場合で、予測値をプロット
lstat = [5, 15, 25, 35]
rm = np.arange(3.50, 8.81, 0.01)

z = lambda beta1, beta2, beta3, rm, lstat: beta1 + beta2*rm + beta3*lstat

plt.scatter(data['RM'],data['y'],label='accutual')
plt.scatter(data['RM'],result.predict(),marker='x',label='predict')

for i in lstat:
  plt.plot(rm, [z(result.params['Intercept'],result.params['RM'],result.params['LSTAT'],r,i) for r in rm],label='lsts={}'.format(i))

plt.legend()
plt.show()
```

```python
mean_squared_error(data['y'],result.predict())
```

```python
result = smf.glm(formula=' y ~ RM+LSTAT',data=data,family=sm.families.Gamma(link=sm.families.links.log())).fit(disp=0)
result.summary()
```

```python
plt.scatter(data['RM'],data['y'],label='accual')
plt.scatter(data['RM'],result.predict(),label='pred')
plt.legend()
plt.show()
```

```python
mean_squared_error(data['y'],result.predict())
```
