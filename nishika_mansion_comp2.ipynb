```python
X['取引時点'].unique()

year = {"年第1四半期": ".25",
        "年第2四半期": ".50",
        "年第3四半期": ".75",
        "年第4四半期": ".99"}
year_list = {}

for i in X["取引時点"].value_counts().keys():
  for k, j in year.items():
    if k in i:
      year_rep = i.replace(k, j)
      year_list[i] = year_rep

year_list
```

```python

X["取引時点"] = X["取引時点"].replace(year_list).astype(float)
```

```python
X['建築年'].unique()
```

```python
y_list = {}
for i in X["建築年"].value_counts().keys():
  if "平成" in i:
    num = float(i.strip("平成").strip("年")[0])
    year = 34 - num
  if "令和" in i:
    num = float(i.strip("令和").strip("年")[0])
    year = 4 - num
  if "昭和" in i:
    num = float(i.strip("昭和").strip("年")[0])
    year = 97 - num
  y_list[i] = year
y_list["戦前"] = 77
X["建築年"] = X["建築年"].replace(y_list)
```

```python
X=X.dropna()
```

```python
X
```

```python
msno.matrix(X)
```

```python
X_train, X_val = train_test_split(X, test_size=0.2)

train_x = X_train.drop('取引価格（総額）_log',axis=1)
train_y = X_train['取引価格（総額）_log']

val_x = X_val.drop('取引価格（総額）_log',axis=1)
val_y = X_val['取引価格（総額）_log']
```

モデル作成

```python
import xgboost as xgb

params = {'objective': 'reg:squarederror',
         'eval_metric': "mae"}

train_data = xgb.DMatrix(train_x, label=train_y)
eval_data = xgb.DMatrix(val_x, label=val_y)

evals = [(train_data, 'train'), (eval_data, 'eval')]

gbm = xgb.train(
    params,
    train_data,
    num_boost_round=100,
    early_stopping_rounds=10,
    evals=evals,
    )
```
