```python
def convert_wareki_to_seireki(wareki):
    if wareki == wareki:
        if wareki == '戦前':
            wareki = '昭和20年'
        value = wareki[2:-1]
        if value == '元':
            value = 1
        else:
            value = int(value)
        if '昭和' in wareki:
            seireki = 1925+value
        elif '平成' in wareki:
            seireki = 1988+value
        elif '令和' in wareki:
            seireki = 2018+value
    else:
        seireki = wareki
    return seireki
```

```python
X['取引時点_何年前'] = X['取引時点'].apply(lambda x: 2021-int(x[:4]))
X.drop(['取引時点'], axis=1, inplace=True)
X['建築年'] = X['建築年'].apply(lambda x: convert_wareki_to_seireki(x))
```
